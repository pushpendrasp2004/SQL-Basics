1. Q1. Create a New Database and  Table for Employees
Task: Create a new database named company_db and Create a table named employees with the following
columns:


Column Name    Data Type       Constraint
employee_id     INT            Primary Key
first_name      VARCHAR(50)
last_name       VARCHAR(50)
department      VARCHAR(50)
salary           INT
hire_date        Date

- SQL Commands: -- Step 1: Create the database
CREATE DATABASE company_db;

-- Step 2: Use the database
USE company_db;

-- Step 3: Create the employees table
CREATE TABLE employees (
    employee_id INT PRIMARY KEY,
    first_name VARCHAR(50),
    last_name VARCHAR(50),
    department VARCHAR(50),
    salary INT,
    hire_date DATE
);

employee_id INT PRIMARY KEY → Ensures each employee has a unique ID.

VARCHAR(50) → Allows text up to 50 characters.

salary INT → Stores whole number salary values.

hire_date DATE → Stores date in YYYY-MM-DD format.

9. What is the Difference Between a Primary Key and a Foreign Key?
- Primary Key                                Foreign Key                Uniquely identifies each record in a table   Links one table to another     Cannot contain NULL values                   Can contain NULL values      
Only one primary key per table              Multiple foreign keys allowed
Ensures entity integrity                   Ensures referential integrity

Example:

employee_id in employees table → Primary Key

If employee_id is used in another table (like payroll), it becomes a Foreign Key there.

10. What Are Constraints in SQL and Why Are They Used?
- Constraints are rules applied to table columns to enforce data integrity and accuracy.

Common Types of Constraints:

a)PRIMARY KEY → Unique + Not NULL

b)FOREIGN KEY → Maintains relationship between tables

c)NOT NULL → Prevents NULL values

d)UNIQUE → Prevents duplicate values

e)CHECK → Ensures condition (e.g., salary > 0)

f)DEFAULT → Sets default value

Constraints Are Used because:
To maintain data accuracy

To prevent invalid data entry

To enforce business rules

To maintain relationships between tables

To improve data reliability
